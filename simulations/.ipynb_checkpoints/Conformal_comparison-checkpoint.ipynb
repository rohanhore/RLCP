```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
# Suppress all warnings
warnings. filterwarnings ('ignore')

from scipy.stats import norm
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from tqdm import tqdm

from conditionalconformal.synthetic_data import generate_cqr_data, indicator_matrix
from conditionalconformal import CondConf
```

```python
def simulation(n,setting):
    X = np.random.normal(0,1,n).astype(np.float32)
    if setting == 1:
        Y = X/2 + np.multiply(np.absolute(np.sin(X)),np.random.normal(0,1,n))
    elif setting == 2:
        Y = X/2 + np.multiply(2*norm.pdf(X,loc=0,scale=1.5),np.random.normal(0,1,n))
        
    return X,Y
```

```python
fig = plt.figure()
fig.set_size_inches(10.5, 6)
fig1 = fig.add_subplot(1,2,1)
X, Y = simulation(1000,1)
fig1.plot(X, Y, '.')

fig2 = fig.add_subplot(1,2,2,sharex = fig1,sharey = fig1)
X, Y = simulation(1000,2)
fig2.plot(X, Y, '.')

plt.show()
```

```python
# nominal level is 0.9
alpha = 0.1

X_seq=np.arange(-2.5, 2.5 + 0.25, 0.25)
n_seq=len(X_seq)
radii = 0.4

n_rep = 50
n = 2000

def local_coverage_bins(eps,setting):
    disc = np.arange(-2.5, 2.5 + eps, eps)

    def phi_fn_groups(x):
        mat = indicator_matrix(x, disc)
        return np.hstack((mat, np.ones((x.shape[0], 1))))

    local_coverage = np.zeros((n_seq,))
    local_ball_id = np.zeros((n, n_seq))

    for iter in tqdm(range(n_rep)):
        X_train, Y_train = simulation(n,setting)
        X_calib, Y_calib = simulation(n,setting)
        X_test, Y_test = simulation(n,setting)
        
        X_train = np.reshape(X_train,(n,1))
        X_calib = np.reshape(X_calib,(n,1))
        X_test = np.reshape(X_test,(n,1))
        
        # fit a linear regression
        reg = LinearRegression().fit(X_train, Y_train)
        
        # score function is residual
        score_fn = lambda x, y : y - reg.predict(x)
        score_inv_fn_ub = lambda s, x : [-np.inf, reg.predict(x) + s]
        score_inv_fn_lb = lambda s, x : [reg.predict(x) + s, np.inf]
    
        phi_fn = phi_fn_groups
        infinite_params = {}
        cond_conf = CondConf(score_fn, phi_fn, infinite_params)
        cond_conf.setup_problem(X_calib, Y_calib)
        
        n_test = len(X_test)
        
        lbs = np.zeros((n_test,))
        ubs = np.zeros((n_test,))
        cov = np.zeros((n_test,))
        
        for i in range(n_test):
            x_t = X_test[i,:]
            res = cond_conf.predict(alpha / 2, x_t, score_inv_fn_lb, exact=True, randomize=True)
            lbs[i] = res[0]
            res = cond_conf.predict(1 - alpha / 2, x_t, score_inv_fn_ub, exact=True, randomize=True)
            ubs[i] = res[1]
            cov[i] = Y_test[i] <= ubs[i] and Y_test[i] >= lbs[i]
    
        for i, x_t in enumerate(X_test):
            for j, x_seq in enumerate(X_seq):
                if np.absolute(x_t-x_seq) <= radii:
                    local_ball_id[i, j] = 1
        local_coverage += np.matmul(cov, local_ball_id)/np.matmul(np.ones((n_test,)), local_ball_id)
    
    return local_coverage/n_rep
```

```python
##setting 1##
#high adaptivity
gensplitCP_local_coverage_11 = local_coverage_bins(5/8,1)
#low adaptivity
gensplitCP_local_coverage_12 = local_coverage_bins(5/4,1)
```

```python
##setting 2##
#high adaptivity
gensplitCP_local_coverage_21 = local_coverage_bins(5/8,2)
#low adaptivity
gensplitCP_local_coverage_22 = local_coverage_bins(5/4,2)
```

```python
cp = sns.color_palette()
sns.set(font="DejaVu Sans")
sns.set_style("whitegrid", {'axes.grid' : False})

fig = plt.figure()
fig.set_size_inches(10.5, 6)
fig1 = fig.add_subplot(1,2,1)
fig1.plot(X_seq, gensplitCP_local_coverage_11,marker='o',lw=1,color=cp[0],label="Gibbs et al.-$\mathcal{F}_1$")
fig1.plot(X_seq, gensplitCP_local_coverage_12,marker='o',lw=1,color=cp[1],label="Gibbs et al.-$\mathcal{F}_2$")
fig1.axhline(y = 0.9, color = 'r', linestyle = '-') 
fig1.legend(loc="lower left")
fig1.set_ylim(0.88,0.93)
fig1.set_xlabel("Feature X", fontsize=14, labelpad=10)
fig1.set_ylabel("coverage", fontsize=14, labelpad=10)

fig2 = fig.add_subplot(1,2,2,sharex=fig1,sharey=fig1)
fig2.plot(X_seq, gensplitCP_local_coverage_21,marker='o',lw=1,color=cp[0],label="Gibbs et al.-$\mathcal{F}_1$")
fig2.plot(X_seq, gensplitCP_local_coverage_22,marker='o',lw=1,color=cp[1],label="Gibbs et al.-$\mathcal{F}_2$")
fig2.axhline(y = 0.9, color = 'r', linestyle = '-') 
fig2.legend(loc="lower left")
fig2.set_ylim(0.88,0.93)
fig2.set_xlabel("Feature X", fontsize=14, labelpad=10)
fig2.set_ylabel("coverage", fontsize=14, labelpad=10)


plt.show()
```
